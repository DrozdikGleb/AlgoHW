```python
from main import *
import matplotlib.pyplot as plt
```

```python
def approximate_points(a, b, c, d):
    x_s = [3 * i / 1000 for i in range(1000)]
    y_s = [rational_func(x, a, b, c, d) for x in x_s]
    return x_s, y_s
```

```python
def show_data(data, points, names):
    x_data = [i for (i, _) in data]
    y_data = [i for (_, i) in data]
    plt.figure(figsize=(20,16))
    plt.scatter(x_data, y_data, color='black')
    for point, name in zip(points, names):
        x_s, y_s = approximate_points(*point)
        plt.plot(x_s, y_s, label=name)
    plt.legend()
```

```python
data = generate_data()
```

```python
nelder = nelder_mead(data)
```

```python
leven = levenberg(data)
```

```python
dif_evol = differential_evolution(data)
```

```python
sim_anneal = simultaneous_anneal(data)
```

```python
show_data(data, [nelder, leven, dif_evol, sim_anneal],
          ["Nelder-Mead", "levenberg", "differential_evolution", "simultaneous_anneal"])
```
